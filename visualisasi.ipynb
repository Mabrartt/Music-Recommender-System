```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:\TEL-U\SEMESTER 6\PROYEK SAINS DATA\streamlit_music2\streamlit_music\dataset.csv')
df.head()
```

```python
artist_song_counts = df['Artist Name(s)'].value_counts().reset_index()
artist_song_counts.columns = ['Artist Name(s)', 'Track Count']

print(artist_song_counts)
```

```python
# Membuat plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Artist Name(s)', y='Track Count', data=artist_song_counts, palette='viridis')

# Menambahkan judul dan label
plt.title('Jumlah Lagu per Artis')
plt.xlabel('Nama Artis')
plt.ylabel('Jumlah Lagu')
plt.xticks(rotation=45)

# Menampilkan plot
plt.show()
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
df['Artist Name(s)'] = df['Artist Name(s)'].astype(str)
```

```python
df = df.drop_duplicates(subset=['Track Name', 'Artist Name(s)'])
```

```python
df.isnull().sum()
```

```python
df.dropna
```

```python
# Gabungkan semua nama artis menjadi satu string
all_artists = ' '.join(df['Artist Name(s)'])
```

```python
# Membuat WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_artists)

# Menampilkan WordCloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
